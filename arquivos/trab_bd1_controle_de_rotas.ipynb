# Relatório 3

### Objetivo: Obter registros com bairros de alto nível de periculosidade e hora do dia com maior índice de roubo.

codigo para obtenção do resultado:

```python
result = pd.read_sql_query("""
            select count (nome_bairro) as indice_assaltos,nome_bairro from boletim
            inner join endereco on (boletim.fk_endereco_cod_end = endereco.cod_end)
            inner join bairro on (endereco.fk_bairro_cod_bairro = bairro.cod_bairro)
            inner join municipio on (bairro.fk_municipio_cod_municipio = municipio.cod_municipio)
            inner join logradouro on (endereco.fk_logradouro_cod_logradouro = logradouro.cod_logradouro) 
            group by nome_bairro,nome_municipio order by indice_assaltos;
                            """,conn)
```

# Relatório 4

### Objetivo: Relatorio contendo origem e destino de mercadorias.

Obeservações: levando em consideração dificuldades encontradas, para entender como relacionar as tabelas, no relatório abaixo usamos o conceito de VIEWs para conseguir apresentar as informações necessarias para o relatório. 

```python

mercadorias = pd.read_sql_query("""

create view origem as 
select cod_origem,concat (tipo_logradouro ,' , ',nome_end ,', ',nome_bairro ,', ',nome_municipio ,'.') as origem from entrega
inner join endereco on (entrega.cod_origem = endereco.cod_end)
inner join bairro on (endereco.fk_bairro_cod_bairro = bairro.cod_bairro)
inner join municipio on (bairro.fk_municipio_cod_municipio = municipio.cod_municipio)
inner join logradouro on (endereco.fk_logradouro_cod_logradouro = logradouro.cod_logradouro); 

create view destino as 
select cod_destino,concat (tipo_logradouro ,' , ',nome_end ,', ',nome_bairro ,', ',nome_municipio ,'.') as destino from entrega
inner join endereco on (entrega.cod_destino = endereco.cod_end)
inner join bairro on (endereco.fk_bairro_cod_bairro = bairro.cod_bairro)
inner join municipio on (bairro.fk_municipio_cod_municipio = municipio.cod_municipio)
inner join logradouro on (endereco.fk_logradouro_cod_logradouro = logradouro.cod_logradouro);  
 
select distinct (cod_entrega),nome,origem.origem,destino.destino from entrega
inner join origem on(origem.cod_origem = entrega.cod_origem)
inner join destino on (destino.cod_destino = entrega.cod_destino)
inner join pessoa on (entrega.FK_CLIENTE_FK_PESSOA_ID = pessoa.id);

                                """,conn)
```

```python
mercadorias
```
